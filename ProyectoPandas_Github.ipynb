### A) Limpia el dataframe de aquellas columnas con todos sus campos a NAN

```python
import pandas as pd
import numpy as np
parking= pd.read_csv("sample_NYC_parking_issues.csv", sep=";")
parking.head()
```

```python

columna_Nan = parking.isna().sum()
print(columna_Nan)
```

```python
Parkin_limpio =parking.fillna(0)
Parkin_limpio
```

### B) Elimina todas las muestras donde la fecha no tenga coherencia: `Vehicle Year`, además transformalas a enteros (2024,2012,2006,etc.)


```python
Parkin_limpio['Vehicle Year'] = Parkin_limpio['Vehicle Year'].astype(int)
Parkin_limpio
```

```python
media_año = Parkin_limpio['Vehicle Year'].replace(0, np.nan).mean()
Parkin_limpio['Vehicle Year'] = Parkin_limpio['Vehicle Year'].replace(0, media_año).fillna(media_año).astype(int)
print(Parkin_limpio['Vehicle Year'].head(5))
```

### C) Hay marcas de vehículos que no guardan sentido con ninguna marca real (`Vehicle Make`) elimina dichas muestras. Deja constancia de cuantas has borrado.

```python
#marcas_unicas = Parkin_limpio['Vehicle Make'].unique()
#print(marcas_unicas)
Parkin_limpio = Parkin_limpio[~Parkin_limpio['Vehicle Make'].isin(['FRIG', 'NS/OT', 'ME/BE', 0, 'FUSO', 'TEAMA', 'PUTEL', 'GIDNY'])]
print(f"Filas eliminadas: {len(Parkin_limpio)}")
```

### D) Representa en un grafíco de barras la cantidad de vehículos (por marca `Vehicle Make`) 

```python

import seaborn as sns
import matplotlib.pyplot as plt

vehiculos_por_marca = Parkin_limpio['Vehicle Make'].value_counts().reset_index()
vehiculos_por_marca.columns = ['Marca', 'Cantidad']
plt.figure(figsize=(12, 8))
#sns.barplot(data=vehiculos_por_marca, x='Cantidad', y='Marca')
Grafico = vehiculos_por_marca.head(10).plot(kind="barh", x='Marca', y='Cantidad', figsize=(12, 8),color='crimson')
plt.title('Cantidad de Vehículos por Marca', fontsize=16)
plt.xlabel('Cantidad', fontsize=12)
plt.ylabel('Marca', fontsize=12)





```
