### A) Limpia el dataframe de aquellas columnas con todos sus campos a NAN

```python
color_marca = color_marca.dropna(subset=['Vehicle Make', 'Vehicle Color', 'Cantidad'])
color_marca = color_marca[color_marca['Vehicle Make'] != 0]
color_marca = color_marca[color_marca['Vehicle Color'] != 0]
color_marca
```

### F) Representa en un gráfico de barras las marca de vehículo segmentadas por colores de vehículos.

```python

color_marca_agrupado = color_marca.groupby(['Vehicle Make', 'Vehicle Color'], as_index=False)['Cantidad'].sum()

top_marcas = color_marca_agrupado.groupby('Vehicle Make')['Cantidad'].sum().nlargest(10).index
top_colores = color_marca_agrupado.groupby('Vehicle Color')['Cantidad'].sum().nlargest(10).index #nlargest(10) para seleccionar los 10 colores con la mayor cantidad.

color_marca_filtrado = color_marca_agrupado[
    (color_marca_agrupado['Vehicle Make'].isin(top_marcas)) & 
    (color_marca_agrupado['Vehicle Color'].isin(top_colores))
]

pivot_data = color_marca_filtrado.pivot(index='Vehicle Make', columns='Vehicle Color', values='Cantidad').fillna(0)

pivot_data.plot(kind='bar', stacked=True, figsize=(15, 9), colormap='tab20')  #pivot reorganizar el DataFrame, stacked=True apilar las barras en el gráfico
plt.title('Distribución de Colores por las Marcas Más Comunes')
plt.xlabel('Marca del Vehículo')
plt.ylabel('Cantidad')
plt.legend(title='Colores')
plt.tight_layout()
plt.show()



```

### I) Convierte la serie `Issue Date` en una serie temporal y contabiliza las multas por `Violation Code[] y més del Issue Date.

```python
#print(parking['Issue Date'].head())  
#print(parking.shape)
parking['Issue Date'] = pd.to_datetime(parking['Issue Date'], format='%m/%d/%Y') #convierte una serie de datos a un formato de fecha y hora 
parking['Year'] = parking['Issue Date'].dt.year
parking['Month'] = parking['Issue Date'].dt.month
multa_conta = parking.groupby(['Violation Code', 'Year', 'Month']).size().reset_index(name='Contabilizada')
multa_conta
```

### J) Visualiza la anterior cuestión (I)

```python
total_multas_año= multa_conta.groupby('Year')['Contabilizada'].sum().reset_index()
total_multas_año
```
